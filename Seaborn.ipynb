**Bar Plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var = sns.load_dataset("penguins")
var
```

```python
sns.set(style="darkgrid")
order_1= ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=var ,hue="sex",order=order_1,
            hue_order=["Female","Male"],palette="icefire",saturation=10,capsize=0.5,errcolor="b")
plt.show()
```

```python
order_1= ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=var,orient="h",color="b")
plt.show()
```

**Pair Plot**

**Box Plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

var = sns.load_dataset("tips")
var
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,hue="sex",showmeans=True
            ,meanprops={"marker":"*","markeredgecolor":"r"},linewidth=2,
            palette="PuOr")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="total_bill",y="day",data=var,hue="sex",showmeans=True
            ,meanprops={"marker":"*","markeredgecolor":"r"},linewidth=2,
            palette="PuOr")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(data=var,showmeans=True
            ,meanprops={"marker":"*","markeredgecolor":"r"},linewidth=2,
            palette="PuOr",orient= "h")
plt.show()
```
